What is Matplotlib?

Matplotlib is a popular Python library for creating static, animated, and interactive visualizations. It offers a wide range of functionalities for:

Plotting various data types: Line charts, scatter plots, bar charts, histograms, pie charts, heatmaps, 3D plots, and more.
Customization: Extensive control over plot elements like colors, markers, lines, axes, labels, titles, legends, and layouts.
Integration: Works seamlessly with other scientific Python libraries like NumPy, pandas, and SciPy for data analysis and manipulation.
Why is Matplotlib Used?

Matplotlib is widely used for various reasons:

Versatility: Creates diverse plots for exploring and understanding data across different domains.
Ease of Use: Provides a user-friendly interface (especially the Pyplot module) for basic and common visualizations.
Customization: Offers a high degree of control for tailoring plots to specific needs and presentations.
Open Source: Freely available and well-documented, with a large community for support.
Integration: Integrates well with other scientific Python tools for a comprehensive data science workflow.

```python
#Q2
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('scatter graph')
plt.scatter(x,y)
plt.show()
```

```python
#Q3
#Purpose of subplot() function:

#Organizes multiple plots within a single figure: Arranges plots in a grid-like structure for visual comparison and efficient space usage.
#Flexible layout options: Combine numerous plots with varying sizes and positions for effective presentations.

import numpy as np
import matplotlib.pyplot as plt

# Data for the four lines
x = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a figure with 2 rows and 2 columns of subplots
fig, axes = plt.subplots(2, 2)

# Plot line 1 in the first subplot
axes[0, 0].plot(x, y1)
axes[0, 0].set_title("Line 1")

# Plot line 2 in the second subplot
axes[0, 1].plot(x, y2)
axes[0, 1].set_title("Line 2")

# Plot line 3 in the third subplot
axes[1, 0].plot(x, y3)
axes[1, 0].set_title("Line 3")

# Plot line 4 in the fourth subplot
axes[1, 1].plot(x, y4)
axes[1, 1].set_title("Line 4")

# Adjust layout and spacing for better visualization
plt.tight_layout()

# Display the plot
plt.show()
```

```python
#q4
#A bar plot is a visualization that uses rectangular bars to represent the values for different categories of data. The length or height of each bar is proportional to the value it represents, allowing for easy comparison between categories.

#Bar plots are commonly used for:

#Comparing values across categories: They provide a clear visual representation of how different categories perform or contribute to a larger whole.
#Visualizing nominal or ordinal data: They are suitable for data where the categories have a specific order or ranking, but not necessarily a meaningful interval between them (e.g., company names, product types).


import numpy as np
import matplotlib.pyplot as plt

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Bar plot
plt.figure(figsize=(10, 6))  # Adjust figure size for better readability
plt.bar(company, profit, color=['red', 'green', 'blue', 'purple'])  # Set colors for bars
plt.xlabel("Company")
plt.ylabel("Profit (in thousands)")
plt.title("Company Profits (Bar Plot)")
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better visibility
plt.tight_layout()
plt.show()

# Horizontal bar plot
plt.figure(figsize=(6, 10))  # Adjust figure size for horizontal bars
plt.barh(company, profit, color=['red', 'green', 'blue', 'purple'])
plt.xlabel("Profit (in thousands)")
plt.ylabel("Company")
plt.title("Company Profits (Horizontal Bar Plot)")
plt.tight_layout()
plt.show()

```

```python
#Q5

#A box plot is a visual representation of the distribution of a numerical dataset.

#Box plots are useful for:

#Comparing the distribution of data across different groups or samples: They provide a compact way to understand central tendency, spread, and potential outliers in each group.
#Identifying potential skewness or outliers: The asymmetry of the box and the presence of data points outside the whiskers can indicate these characteristics.


import numpy as np
import matplotlib.pyplot as plt

# Generate sample data
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Create the box plot
plt.figure(figsize=(8, 6))
plt.boxplot([box1, box2], labels=["Box 1", "Box 2"], notch=True, vert=True, patch_artist=True)
plt.xlabel("Data Group")
plt.ylabel("Value")
plt.title("Box Plot Comparison")
plt.grid(True)  # Add grid lines for better readability
plt.tight_layout()
plt.show()
```
