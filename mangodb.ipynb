Q1.MongoDB is a document-oriented, open-source NoSQL database. It's designed for flexible, scalable, and high-performance data storage, particularly for handling large volumes of unstructured or semi-structured data.
Non-relational databases (NoSQL) depart from the traditional relational database model (RDBMS)
Data structures: NoSQL databases employ various data structures beyond tables (e.g., documents, key-value pairs, graphs), each suited for specific use cases.
Schema flexibility: Often schema-free or offer flexible schema definitions, accommodating evolving data structures.

Scenarios where MongoDB is preferred over SQL databases:

Unstructured or semi-structured data: Excels at handling data that doesn't fit neatly into predefined schemas, such as product catalogs, user profiles, logs, or sensor data.
Rapid development and frequent schema changes: Adapts quickly to evolving data requirements without complex schema migrations.
High scalability and performance: Handles large-scale data and high-volume traffic efficiently.
